```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Bank Data.csv')
```

```python
# Display the first few rows and basic information
print(df.head())
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handle missing values
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Explore distribution of key numerical features
key_numerical_features = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 
                          'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date', 
                          'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Num_Credit_Inquiries', 
                          'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Credit_History_Age', 
                          'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance']

plt.figure(figsize=(20, 30))
for i, feature in enumerate(key_numerical_features, 1):
    plt.subplot(6, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Check for outliers using box plots
plt.figure(figsize=(20, 10))
df[key_numerical_features].boxplot()
plt.title('Box Plots to Detect Outliers')
plt.xticks(rotation=90)
plt.show()
```

```python
# Handle outliers for key features
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.clip(df[column], lower_bound, upper_bound)

for feature in key_numerical_features:
    handle_outliers(df, feature)
```

```python
# Encode categorical variables
categorical_columns = ['Occupation', 'Type_of_Loan', 'Credit_Mix', 'Payment_Behaviour']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
# Create Credit_Score target variable
def assign_credit_score(row):
    score = 0
    score += min(row['Annual_Income'] / 10000, 30)  # Up to 30 points for income
    score += min(row['Credit_History_Age'] / 12, 20)  # Up to 20 points for credit history age
    score += max(10 - row['Num_of_Delayed_Payment'], 0) * 2  # Up to 20 points for fewer delayed payments
    score += min(row['Num_Bank_Accounts'], 5) * 2  # Up to 10 points for bank accounts
    score += min(row['Num_Credit_Card'], 3) * 5  # Up to 15 points for credit cards
    score += max(0, min(80 - row['Credit_Utilization_Ratio'], 20))  # Up to 20 points for lower utilization
    return score

df['Credit_Score'] = df.apply(assign_credit_score, axis=1)
```

```python
# Categorize Credit_Score
df['Credit_Score_Category'] = pd.cut(df['Credit_Score'], 
                                     bins=[0, 40, 60, 80, np.inf], 
                                     labels=['Poor', 'Standard', 'Good'])
```

```python
# Explore distribution of target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='Credit_Score_Category', data=df)
plt.title('Distribution of Credit Scores')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(df[key_numerical_features + ['Credit_Score']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot for key features
key_features = ['Annual_Income', 'Credit_Utilization_Ratio', 'Credit_History_Age', 'Credit_Score']
sns.pairplot(df[key_features], hue='Credit_Score_Category')
plt.show()

print(df['Credit_Score_Category'].value_counts(normalize=True))
```
